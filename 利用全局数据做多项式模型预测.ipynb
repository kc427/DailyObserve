```python
import pandas as pd
import numpy as np
data = pd.read_excel('school_enroll_unit_admission.xlsx')
data.head(5) #读取数据
```

```python
data2 = data[['enroll_year','school_id','min_score']]
data2.head()
```

```python
import matplotlib.pyplot as plt  #引入绘图库
from sklearn.model_selection import train_test_split #数据拆分
from sklearn.linear_model import LinearRegression
```

```python
data4 =data2.drop_duplicates(subset=['enroll_year','school_id'], keep='first', inplace=False)
```

```python
index_series = data4.set_index([ "school_id","enroll_year"])["min_score"]
new_df = index_series.unstack()
new_df.isnull().sum()
```

```python
new_df.head()
```

```python
new_df = new_df.rename_axis(columns=None)
new_df = new_df.reset_index()
new_df.head()
```

```python
new_df=new_df.dropna(axis=0,how='any')
new_df=new_df.drop(new_df[new_df[2020]<=300].index)
new_df=new_df.drop(new_df[new_df[2017]<=300].index)
new_df=new_df.drop(new_df[new_df[2018]<=300].index)
new_df=new_df.drop(new_df[new_df[2019]<=300].index)
```

```python
x=new_df.iloc[:,1:4]
x2 = x.values.reshape(-1,1)
y=new_df.iloc[:,4:].values.reshape(-1,1)
x2
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.25)
```

```python
linreg2 = LinearRegression() #调用最小2乘法的模型
linreg2.fit(X_train,  y_train)
print(linreg2.coef_)
print(linreg2.intercept_)
#模型训练过程
```

```python
linreg2.score(X_test,y_test) #模型准确率96%
```

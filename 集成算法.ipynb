```python

def enroll_year(list,data):
    dict ={}
    for  i  in list:
    
        data3 =data[data['school_id']==i]
    
        x = data3['enroll_year'].values.reshape(-1, 1)
        y = data3['min_score_diff']
        linreg = LinearRegression()
        linreg.fit(x, y)
        dict[i]=linreg.predict([[2021]])
    return dict
    #linreg2.append(linreg.coef_)
    #intercept.append(linreg.intercept_)
    #dict[2021] =linreg.predict([[2021]])
```

```python
# 预测2021的分数

def minscore(list,data):
    dict ={}
    for  i  in list:
    
        data3 =data[data['school_id']==i]
    
        x = data3['enroll_year'].values.reshape(-1, 1)
        y = data3['min_score']
        linreg = LinearRegression()
        linreg.fit(x, y)
        dict[i]=linreg.predict([[2021]])
    return dict
    #linreg2.append(linreg.coef_)
    #intercept.append(linreg.intercept_)
    #dict[2021] =linreg.predict([[2021]])
```

```python
# 预测2021的分数线

def tidangscore(list,data):
    dict ={}
    for  i  in list:
    
        data3 =data[data['school_id']==i]
    
        x = data3['enroll_year'].values.reshape(-1, 1)
        y = data3['score_tidang']
        linreg = LinearRegression()
        linreg.fit(x, y)
        dict[i]=linreg.predict([[2021]])
    return dict
    #linreg2.append(linreg.coef_)
    #intercept.append(linreg.intercept_)
    #dict[2021] =linreg.predict([[2021]])
```

```python
dict_gaozhi_1=enroll_year(list_gaozhi_1,data_gaozhi_1)
dict_gaozhi_2=enroll_year(list_gaozhi_2,data_gaozhi_2)
dict_erpi_1=enroll_year(list_erpi_1,data_erpi_1)
dict_erpi_2=enroll_year(list_erpi_2,data_erpi_2)
dict_yipi_1=enroll_year(list_yipi_1,data_yipi_1)
dict_yipi_2=enroll_year(list_yipi_2,data_yipi_2)
```
